```python


# Definindo a senha alvo e os caracteres permitidos
clave_deseada = "pucCN2024!"
alfabeto = " abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!.@#$%&1234567890"
```

```python
len(alfabeto)
```

```python
import random
import datetime

# Função para gerar um indivíduo (uma possível senha) aleatoriamente
def generar_clave_aleatoria(longitud):
    genes = []
    while len(genes) < longitud:
        sampleSize = min(longitud - len(genes), len(alfabeto))
        genes.extend(random.sample(alfabeto, sampleSize))
    return ''.join(genes)

# Função para calcular o fitness (quão próxima a senha está da senha alvo)
def calcular_aptitud(clave_intento):
    return sum(1 for esperado, actual in zip(clave_deseada, clave_intento) if esperado == actual)

# Função para realizar a mutação em um indivíduo
def mutar_clave(clave_padre):
    index = random.randrange(0, len(clave_padre))
    genes_hijo = list(clave_padre)
    nuevo_gen, alternativo = random.sample(alfabeto, 2)
    if genes_hijo[index] == nuevo_gen:
        genes_hijo[index] = alternativo
    else:
        genes_hijo[index] = nuevo_gen
    return ''.join(genes_hijo)
```

```python
generar_clave_aleatoria(80)
```

```python
# Função para exibir o resultado
def mostrar_resultado(clave_intento):
    tiempo_transcurrido = datetime.datetime.now() - inicio_tiempo
    aptitud = calcular_aptitud(clave_intento)
    print("\n{0}\t{1}\t{2}".format(clave_intento, aptitud, str(tiempo_transcurrido)))
```

```python
# Inicializando o processo
random.seed()
inicio_tiempo = datetime.datetime.now()
mejor_clave = generar_clave_aleatoria(len(clave_deseada))
mejor_aptitud = calcular_aptitud(mejor_clave)
mostrar_resultado(mejor_clave)

# Loop principal do algoritmo genético
while True:
    clave_hijo = mutar_clave(mejor_clave)
    aptitud_hijo = calcular_aptitud(clave_hijo)
    mostrar_resultado(clave_hijo)

    # Medir la longitud de cada gen (cada caracter en la clave)
    longitud_genes = [len(gen) for gen in clave_hijo]
    print(f"Longitud de cada gen en clave_hijo: {longitud_genes}")

    if mejor_aptitud >= aptitud_hijo:
        continue
    if aptitud_hijo >= len(mejor_clave):
        break
    mejor_aptitud = aptitud_hijo
    mejor_clave = clave_hijo

# Mostrar la mejor contraseña generada y su aptitud
print(f"\nLa mejor contraseña generada es: '{mejor_clave}' con una aptitud de {mejor_aptitud}")
```
